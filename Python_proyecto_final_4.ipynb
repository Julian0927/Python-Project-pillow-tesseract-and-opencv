En OpenCV `cvtColor` es una función de la librería OpenCV que permite cambiar el espacio de color de una imagen. Por ejemplo, el espacio de color se refiere a la representación de  cada canal de una imagen, por ejemplo RGB, SRGB, BGR, CMYK, etc. La función, requiere 3 parámetros de entrada los cuales son:

$ \ $

*   ***Parámetro(1):*** Una imágen en un espacio de colores determinado (RGB,SRGB, BGR, CMYK,..., etc). Esta imagen está representada por una matriz.

$ \ $

*   ***Parámetro(2):*** Una imágen en un espacio de colores determinado (RGB,SRGB,BGR, CMYK,..., etc). Esta imagen está representada por una matriz.

$ \ $

*   ***Parámetro(3):*** Código de conversión.

$ \ $

Y tiene una salida dada por:

$ \ $


*   ***Salida:*** Aparentemente es una imagen del parámetro (2), la cual está descrita por medio de una matriz. Esta imagen se obtiene gracias al código de conversión y la imagen del parámetro (1).

$ \ $


```python
#(1) importamos el módulo cv2 y lo denotamos de ahora en adelante como cv.
import cv2 as cv


#(2) usamos la función cvtColor del módulo cv para convertir a escala de grises la imagen dada  usando como código de conversión cv.COLOR_BGR2GRAY.
#parámetro(1) = imagen_matricial
#parámetro(3) = cv.COLOR_BGR2GRAY <------->  convertiremos a escala de grises usando la imagen dada y el código cv.COLOR_BGR2GRAY (para escala de grises)
gris = cv.cvtColor(imagen_matricial, cv.COLOR_BGR2GRAY)


#(3) mostramos en pantalla la representación matricial de la imagen convertida a escala de grises llamada gris.
# print(gris) ≈ display(gris).
print(gris)

```

La función `reshape` da un nuevo tamaño a la matriz sin cambiar los datos de esta. La función se describe como:

$$\text{reshape(Matriz, Nuevo tamaño, order='C')}$$

donde los parámetros de `reshape` son:


**(a) Parámetro (1):** Matriz ------------> matriz a reformar.

**(b) Parámetro (2):** Nuevo tamaño ----------> entero o una tupla de enteros.

El nuevo tamaño de la matriz debe ser compatible con el tamaño de la matriz original.  

**(c) Parámetro (3):** order : {'C', 'F', 'A'}, optional. Este parámetro lee los elementos los coeficientes de la matriz del parámetro (1)  y coloca los elementos en la matriz remodelada utilizando este orden de índice. 


*  'C'  significa leer/escribir los elementos usando un orden de índice similar al de `C`, con el índice del último eje cambiando más rápido, hasta volver al primer índice del eje cambiando de una  manera más lenta.


* 'F' significa leer/escribir los elementos usando un orden de índice similar a `Fortran`, con el primer índice cambiando más rápido y el último índice cambiando más lento. 


* 'A' significa leer/escribir los elementos de manera similar al orden de indice de `Fortran`, si la matriz del parámetro (1) es "contiguous" `Fortran` y en caso contrario  leer/escribir los elementos de manera similar al orden de indice de `C`.



**Salida**= matriz remodelada.

$ \ $

```python
#(1) importamos la librería de imagenes PIL.
import PIL


#(2) importamos el módulo Image de la librería de imagenes PIL.
from PIL import Image


#(3) importamos la librería numpy.
import numpy as np


#(4) importamos el paquete cv2. Este paquete representa básicamente a OpenCV.
import cv2 as cv


#(5) cargamos la ruta "floyd.jpg" y le damos el nombre de imagen_matricial.
imagen_matricial = cv.imread("floyd.jpg")


#(6) creamos una copia de imagen_matricial en escala de grises y le damos el nombre de gris.
gris = cv.cvtColor(imagen_matricial, cv.COLOR_BGR2GRAY)


#(7) mostramos en pantalla el resultado del literal (6).
print(gris)

```

```python
#(8) describimos el tamaño 1x(m*n) de la imagen matricial con nombre gris usando la función shape.
(m,n)=gris.shape


#(9) describimos una nueva variable llamada nuevo_tam que representa una tupla de números. 
nuevo_tam=(1,m*n)


#(10) usamos la función reshape de la librería numpy para construir una nueva matriz a partir de la matriz gris y la tupla nuevo_tam.
gris_1=np.reshape(gris,nuevo_tam)


#(11) mostramos en pantalla el resultado de (10).
print(gris_1)

```

```python
#(12) tomamos la matriz gris_1 y describimos la imagen que representa en la variable imagen_gris_rara. Es claro que esta imagen no se va a ver debido a que la matriz que representa esta imagen es de tamaño 1x(m*n)
imagen_gris_rara = Image.fromarray(gris_1, "L")


#(13) mostramos en pantalla la imagen construida en (12).
display(imagen_gris_rara)

```
