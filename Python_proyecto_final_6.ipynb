Una de las cosas buenas de usar los sistemas de portátiles en Jupyter/colab es que hay un abundante conjunto de complementos aportados que buscan extender este sistema. En esta parte veremos uno de estos complementos, el cual es llamado `ipywebrtc`.  Para ser más precisos,  `ipywebrtc` es un protocolo para comunicación en tiempo real en la web, de hecho es lo que conocemos como ***chatear***. 


 Vamos a ver. Primero, importemos de esta biblioteca dos clases diferentes que usaremos en un demo, uno para la cámara y otro para las imágenes.

```python
#(1) primero instalamos ipywebrtc
!pip install ipywebrtc
```

```python
#(2) importamos dos clases que usaremos "CameraStream" para la cámara y "ImageRecorder" para las imágenes
from ipywebrtc import CameraStream, ImageRecorder
```

```python
#(3) miramos la documentación de la clase CameraStream
help(CameraStream)
```

```python
#(4) miramos la documentación de la clase ImageRecorder
help(ImageRecorder)
```

```python
camara = CameraStream.facing_user(audio=False)
```

```python
image_recorder = ImageRecorder(stream=camara)

image_recorder.recording=True

image_recorder.download()

type(image_recorder.image)
```
